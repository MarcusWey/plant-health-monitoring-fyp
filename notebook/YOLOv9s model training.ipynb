```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/drive/MyDrive/Marcus Colab/marcus_plant_disease_detection/yolo_plant_detection/YOLOv9/yolov9s_plant_model/results.csv")

epochs = df.index + 1
map50 = df['metrics/mAP50(B)']
map5095 = df['metrics/mAP50-95(B)']

plt.figure(figsize=(10, 6))
plt.plot(epochs, map50, 'o-', label='mAP@0.5')
plt.plot(epochs, map5095, 'o-', label='mAP@0.5:0.95')
plt.xlabel("Epochs")
plt.ylabel("mAP")
plt.title("mAP Over Epochs")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("real_mAP_plot.png")
plt.show()

!mv real_mAP_plot.png "/content/drive/MyDrive/Marcus Colab/marcus_plant_disease_detection/yolo_plant_detection/YOLOv9/yolov9s_plant_model/"
```
